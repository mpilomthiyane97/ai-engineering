```python
# Step 1: Install required packages
# - sentence-transformers: for easy sentence embeddings
# - scikit-learn: for cosine similarity and PCA
# - seaborn: for clean visual plots
!pip install sentence-transformers scikit-learn seaborn

# Step 2: Import necessary libraries
from sentence_transformers import SentenceTransformer           # For loading the model
from sklearn.metrics.pairwise import cosine_similarity           # To measure similarity between sentence vectors
from sklearn.decomposition import PCA                           # To reduce embedding dimensions for visualization
import matplotlib.pyplot as plt                                 # For plotting graphs
import seaborn as sns                                           # For prettier plot styles
import numpy as np                                              # For numerical operations
import pandas as pd                                             # For showing similarity matrix in table form

# Step 3: Load a pre-trained sentence embedding model
# This model converts full sentences into 384-dimensional vectors
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 4: Define a list of sentences to compare
sentences = [
    "I love dogs",                          # semantically similar to sentence 2
    "I adore puppies",                      # similar meaning to sentence 1
    "The sun is bright",                    # semantically unrelated
    "He enjoys playing with his dog",       # somewhat related to 1 and 2
    "I dislike thunderstorms"               # emotionally opposite to sentence 1
]

# Step 5: Encode the sentences into vector embeddings
# The output is a list of 384-dimensional numpy arrays
embeddings = model.encode(sentences)

# Step 6: Print part of the vector for each sentence to inspect it
for sentence, vec in zip(sentences, embeddings):
    print(f"\nSentence: {sentence}")
    print(f"Vector (first 5 dims): {vec[:5]}")  # only showing first 5 dimensions for readability

# Step 7: Compute cosine similarity between every pair of sentence vectors
# Higher values mean more similar meaning (range: -1 to 1)
sim_matrix = cosine_similarity(embeddings)

# Step 8: Reduce the 384-dimensional vectors down to 2D using PCA (for plotting)
pca = PCA(n_components=2)                   # Only keep 2 principal components
reduced = pca.fit_transform(embeddings)     # Now each sentence has 2D coordinates

# Step 9: Plot the 2D coordinates of each sentence
plt.figure(figsize=(10, 6))                 # Set the plot size
sns.set(style="whitegrid")                  # Optional: nicer grid style

# Plot each sentence as a point with a text label
for sent, (x, y) in zip(sentences, reduced):
    plt.scatter(x, y)                       # plot the point
    plt.text(x + 0.01, y + 0.01, sent, fontsize=12)  # label the point with the sentence

# Add titles and axis labels
plt.title("2D PCA of Sentence Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()

# Step 10: Display cosine similarity matrix as a table
# Shows how semantically similar each sentence is to the others
print("\nCosine Similarity Matrix:")
display(pd.DataFrame(sim_matrix, index=sentences, columns=sentences).round(2))
```
